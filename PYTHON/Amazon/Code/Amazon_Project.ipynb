```python
driverPath=r"C:\Users\Merit\Downloads\chromedriver_win32\chromedriver.exe"
driver = webdriver.Chrome(driverPath)
driver.get("https://www.amazon.in/s?k=offers+washing+machine&crid=2GJ3Q0TXHIXDV&qid=1629821109&sprefix=offers+washin%2Caps%2C307&ref=sr_pg_1")
writer = pd.ExcelWriter("Amazon_product.xlsx", engine='xlsxwriter')
Description = []
Price = []
Offers = []
for x in range(0,20):
    description=driver.find_elements_by_xpath("//span[@class='a-size-medium a-color-base a-text-normal']")
    for i in description:
        Description.append(i.text)
    price=driver.find_elements_by_xpath("//span[@class='a-price a-text-price']")
    for j in price:
        Price.append(j.text)
    offer=driver.find_elements_by_xpath("//span[@class='a-price-whole']")
    for k in offer:
        Offers.append(k.text)
    driver.find_element_by_xpath("//li[@class='a-normal'][2]").click()
    time.sleep(3)
driver.close()
```

```python
data = pd.DataFrame(list(zip(Description,Price,Offers)),columns=["Description","Price","Offers"])
data.to_excel(writer,sheet_name='washing_machine',index = False)
writer.save()
```
